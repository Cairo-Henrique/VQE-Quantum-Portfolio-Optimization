## 1. Introdução: Problema de Otimização Binária de Portfólios Financeiros

Neste trabalho, aborda-se a otimização binária de portfólios financeiros como um problema QUBO, no qual variáveis binárias indicam a inclusão ou não de ativos na carteira, visando maximizar o retorno esperado e minimizar o risco sob um certo conjunto de restrições (por exemplo, orçamento fixo). Diferentemente da abordagem clássica da Teoria Moderna de Portfólio, de Markowitz, o problema é mapeado para um Hamiltoniano e resolvido por meio do *Variational Quantum Eigensolver* (VQE), permitindo um estudo comparativo entre métodos clássicos (Markowitz) e híbridos quântico-clássicos (VQE). Na abordagem do VQE, utiliza-se um circuito quântico parametrizado (*ansatz*) para aproximar o estado de menor energia, correspondente à melhor seleção de ativos. Analisamos de forma comparativa ambas as soluções a fim de entender a importância de explorar métodos alternativos para aprimorar a tomada de decisões financeiras, viabilizando estratégias mais eficientes e inovadoras no mercado.


## 2. Metodologia

O problema é formulado como uma otimização com restrição de cardinalidade, na qual o investidor pode selecionar no máximo quatro ativos dentre os dez disponíveis, caracterizando um problema combinatório. Essa formulação é resolvida por duas abordagens:

- **Clássica**: baseada na Teoria Moderna de Portfólio, de Harry Markowitz, com pesos contínuos;
- **Híbrida (quântico-clássica)**: utilizando o algoritmo VQE, que modela o problema como um QUBO com variáveis binárias.

A solução do problema é conduzida por meio de duas abordagens complementares. Na abordagem clássica, utiliza-se um processo iterativo de otimização que ajusta os parâmetros até a convergência da função objetivo. Em paralelo, na abordagem híbrida quântico-clássica, o problema QUBO é mapeado para um Hamiltoniano e resolvido utilizando o algoritmo variacional, no qual um circuito quântico parametrizado (*ansatz*) é otimizado por um otimizador clássico. Essa estratégia permite explorar a minimização de energia do sistema como análogo direto à otimização do portfólio, possibilitando a comparação entre métodos clássicos e quânticos em termos de desempenho e qualidade da solução.

O objetivo, em ambos os casos, é encontrar o melhor equilíbrio entre retorno esperado e risco, controlado por um parâmetro de aversão ao risco, permitindo comparar o desempenho e as características das soluções obtidas por métodos tradicionais e quânticos.

## 4. Teoria Moderna do Portfólio de Markowitz

Na abordagem clássica, baseada na Teoria Moderna do Portfólio de Harry Markowitz, o problema é formulado como uma otimização contínua, na qual os pesos dos ativos $ w_i \in [0,1] $ representam a fração do capital investido em cada ativo. A função objetivo a ser minimizada combina dois componentes principais:

1. Termo de risco proporcional à matriz de covariância dos ativos, ponderado por um parâmetro de aversão ao risco ( $q$ );
2. Termo linear associado ao retorno esperado dos ativos.

Formalmente, a função objetivo pode ser expressa como:

$$
f(w) = q \sum_{i,j} \sigma_{ij} w_i w_j - \sum_{i=1}^{N} \mu_i w_i,
\tag{3}
$$

sujeita à restrição de orçamento:

$$
\sum_{i=1}^{N} w_i = 1,
\tag{4}
$$

onde:

- $w_i$ representam os pesos contínuos dos ativos no portfólio;
* $\sigma_{ij}$ é a matriz de covariância;
* $\mu_i$ são os retornos esperados;
* $q$ é o parâmetro de aversão ao risco.

Essa formulação permite determinar a alocação ótima de capital entre os ativos, equilibrando risco e retorno de forma contínua, servindo como base para comparação com a abordagem discreta baseada em QUBO.


### 4.1 Fronteira Eficiente de Markowitz

Por meio de simulações de Monte Carlo, foram gerados milhares de portfólios com pesos contínuos aleatórios $w$, respeitando $\sum_{i=1}^{N} w_i = 1$.

Para tornar a análise mais robusta, diferentes métodos de estimação foram utilizados para os retornos esperados, matriz de covariância e medidas de risco. Em particular, foram considerados os seguintes estimadores:

- Média histórica (Mean) — estimativa tradicional baseada na média amostral;
- Média móvel exponencial (EWMA) — atribui maior peso para observações mais recentes;
- Janela deslizante (Rolling window) — utiliza apenas uma janela temporal fixa mais recente;
- Estimador de Ledoit–Wolf — regularização da matriz de covariância para reduzir erro de estimação.

Para cada portfólio, foram calculados:

- **Retornos dos ativos** $R_i$ (percentuais):
> - `returns = data.pct_change().dropna()`

- **Retorno esperado anualizado (252 dias úteis):**
> - `estimate_return(returns, method="mean", window=60, span=60)`

$$
\begin{align}
\mu_i &= \mathbb{E}[R_i] \cdot 252\,, \tag{5} \\
R_p &= w^T \mu\,.
\end{align}
$$


- **Matriz de covariância anualizada:**
> - `estimate_covariance(returns, method="mean", window=60, span=60)`

$$
\Sigma = \text{Cov}(R) \cdot 252\,.
\tag{6}
$$

- Risco (volatilidade):  
> - `portfolio_risk(weights, returns_df, sigma, risk_measure="variance", alpha=0.05)`

Foram consideradas as seguintes medidas:

- Variância (Volatilidade)
- Value-at-Risk (VaR)
- Conditional Value-at-Risk (CVaR)

A volatilidade do portfólio é definida por:

$$
\sigma_p = \sqrt{w^T \Sigma w}\,.
\tag{7}
$$

Para as métricas VaR e CVaR, o risco foi estimado diretamente a partir da distribuição empírica dos retornos do portfólio.

- Escore (índice de Sharpe modificado):  
$$
S = \frac{R_p}{\sigma_p}\,.
\tag{8}
$$

Essa análise permite:

- Visualizar o espaço risco-retorno;
- Identificar o portfólio ótimo clássico (máximo escore);
- Estabelecer um **limite superior de eficiência**.


Esses parâmetros são fundamentais para a formulação do problema de otimização, pois:

- $\mu$ representa o **componente de retorno**;
- $\Sigma$ representa o **componente de risco e dependência**.

Eles serão posteriormente utilizados tanto no modelo clássico quanto na formulação QUBO para o algoritmo quântico.

```python
def estimate_return(returns, method="mean", window=60, span=60):
    if method == "mean":
        mu = returns.mean().values * 252
    elif method == "ewma":
        mu = returns.ewm(span=span).mean().iloc[-1].values * 252
    elif method == "rolling":
        mu = returns.tail(window).mean().values * 252
    elif method == "ledoit":
        # Ledoit não afeta retorno — apenas covariância
        mu = returns.mean().values * 252
    else:
        raise ValueError("Método inválido")
    return mu


def estimate_covariance(returns, method="mean", window=60, span=60):
    returns = returns.replace([np.inf, -np.inf], np.nan).dropna()

    if returns.empty:
        raise ValueError("O DataFrame de retornos está vazio após limpeza.")

    if method == "mean":
        sigma = returns.cov().values * 252
    elif method == "ewma":
        sigma = returns.ewm(span=span).cov().iloc[-len(returns.columns):].values * 252
    elif method == "rolling":
        sigma = returns.tail(window).cov().values * 252
    elif method == "ledoit":
        lw = LedoitWolf()
        lw.fit(returns.values)
        sigma = lw.covariance_ * 252
    else:
        raise ValueError("Método inválido")

    return sigma

def portfolio_risk(weights, returns_df, sigma, risk_measure="variance", alpha=0.05):
    if risk_measure == "variance":
        return np.sqrt(np.dot(weights.T, np.dot(sigma, weights)))

    port_daily_returns = returns_df.values @ weights
    var = np.quantile(port_daily_returns, alpha)

    if risk_measure == "var":
        return abs(var) # ou -var
    elif risk_measure == "cvar":
        cvar = port_daily_returns[port_daily_returns <= var].mean()
        return abs(cvar) # ou -cvar
    else:
        raise ValueError("risk_measure inválido")
```

```python
# =========================================================
# Escolha dos métodos
# =========================================================
method = "ledoit"        # "mean", "ewma", "rolling", "ledoit"
risk_measure = "cvar"    # "variance", "var", "cvar"

alpha = 0.05             # Para VaR e CVaR
window = 60              # Para rolling
span = 60                # Para EWMA
risk_free_rate = 0.0

if risk_measure == "variance":
    metric_name = "Sharpe"
elif risk_measure == "var":
    metric_name = "Return / VaR"
elif risk_measure == "cvar":
    metric_name = "Return / CVaR"

# =========================================
# Retorno e Risco
# =========================================

returns = data.pct_change().dropna()
mu = estimate_return(returns, method, window, span)
sigma = estimate_covariance(returns, method, window, span)

print("\n" + "="*50)
print("Parâmetros Clássicos Extraídos")
print("="*50)

print(f"\nAtivos selecionados: {tickers}")

print("\nRetornos Esperados (mu):")
for t, m in zip(tickers, mu):
    print(f"{t}: {m:.4f}")

print("\nMatriz de Covariância (sigma):")
print(sigma)
```

```python
# =========================================================
# Simulação de portfólios
# =========================================================

n_portfolios = 10000
n_assets = len(tickers)

results = np.zeros((3, n_portfolios))
weights_record = []

for i in range(n_portfolios):

    weights = np.random.random(n_assets)
    weights /= np.sum(weights)

    weights_record.append(weights)

    portfolio_return = np.dot(weights, mu)

    portfolio_volatility = portfolio_risk(
        weights,
        returns,
        sigma,
        risk_measure=risk_measure,
        alpha=alpha
    )

    metric = portfolio_return / portfolio_volatility

    results[0,i] = portfolio_return
    results[1,i] = portfolio_volatility
    results[2,i] = metric

#=========================================================
# Melhor portfólio
#=========================================================
max_metric_idx = np.argmax(results[2])

optimal_weights = weights_record[max_metric_idx]
optimal_return = results[0,max_metric_idx]
optimal_vol = results[1,max_metric_idx]

# =========================================================
# Fronteira eficiente
# =========================================================

plt.figure(figsize=(10,6))

plt.scatter(
    results[1,:],
    results[0,:],
    c=results[2,:],
    cmap="viridis",
    s=5
)

plt.colorbar(label=f"{metric_name} Ratio")

plt.scatter(
    optimal_vol,
    optimal_return,
    color="red",
    s=200,
    label="Optimal Portfolio"
)

plt.xlabel("Volatility (Risk)")
plt.ylabel("Expected Return")
plt.title("Efficient Frontier")
plt.legend()
plt.grid()
plt.show()

# =========================================================
# Portfólio Ótimo (Monte Carlo)
# =========================================================
weights_df = pd.DataFrame({
    "Ticker": tickers,
    "Weight": optimal_weights
})


print("\n" + "="*50)
print(f"Portfólio Ótimo (Máximo {metric_name})")
print("="*50)

print("\nPesos ótimos:")
print(weights_df.to_string(index=False))

print("\nMétricas do portfólio:")
print(f"Retorno Esperado: {optimal_return:.4f}")
print(f"Volatilidade:     {optimal_vol:.4f}")
print(f"{metric_name} Ratio:     {(optimal_return / optimal_vol):.4f}")
```

## 5. Algoritmos Híbridos (Quântico-Clássicos)

Para resolver o problema de otimização de portfólio no contexto da computação quântica atual (NISQ), são utilizados **algoritmos híbridos**, nos quais um circuito quântico parametrizado é otimizado por um algoritmo clássico. Esses métodos são particularmente adequados para problemas formulados como **QUBO (Quadratic Unconstrained Binary Optimization)**, como é o caso do portfólio binário adotado neste projeto. A seguir, são descritas três abordagens amplamente utilizadas na literatura.

### 5.1 QAOA (*Quantum Approximate Optimization Algorithm*)

O QAOA é um algoritmo voltado para problemas de otimização combinatória, que alterna operadores quânticos de custo e mistura para gerar estados parametrizados. Seus parâmetros são otimizados por métodos clássicos, buscando minimizar a função objetivo. No contexto do portfólio, o problema QUBO é convertido em um Hamiltoniano de Ising, e o QAOA encontra a melhor combinação binária de ativos.

### 5.2 VQE (*Variational Quantum Eigensolver*)

O VQE é uma abordagem variacional que busca o estado de menor energia de um Hamiltoniano, utilizando circuitos parametrizados (*ansatz*) e otimização clássica. No problema de portfólio, a formulação QUBO é mapeada para um Hamiltoniano, e a solução ótima corresponde ao estado fundamental, representando a melhor seleção de ativos. Trata-se de um método mais geral e flexível.

### 5.3 QNN (Quantum Neural Networks)

As QNNs utilizam circuitos quânticos como modelos de aprendizado, sendo treinadas com algoritmos clássicos. Em vez de resolver diretamente a otimização, são aplicadas para previsão de retornos ou identificação de padrões de mercado, fornecendo insumos para modelos de otimização posteriores.

## 6. VQE Aplicado à Otimização Binária de Portfólios Financeiros

A partir da investigação realizada, foi selecionado o **VQE** para resolver o problema de otimização de portfólio. A escolha se justifica por:

- Ser um algoritmo **robusto na era NISQ (Noisy Intermediate-Scale Quantum)**;
- Exigir **circuitos de baixa profundidade**, reduzindo efeitos de ruído;  
- Permitir integração eficiente com **otimizadores clássicos** (ex: COBYLA).

Para sua aplicação, o problema de portfólio é reformulado como um **Hamiltoniano de Ising**, onde:

- A função de custo financeira (risco-retorno) é mapeada para energia (Hamiltoniano);  
- A solução ótima corresponde ao **estado fundamental (menor energia)**.

### 6.1 Função de custo - Modelagem QUBO (*Quadratic Unconstrained Binary Optimization*)
> - `PortfolioOptimization`

A função de custo (função objetivo) a ser minimizada combina três componentes principais:

1. Termo de risco proporcional à matriz de covariância dos ativos, ponderado por um parâmetro de aversão ao risco ( $q$ );
2. Termo linear associado ao retorno esperado dos ativos;
3. Termo de penalização que garante o atendimento da restrição orçamentária por meio de variáveis auxiliares binárias. Formalmente, a função objetivo pode ser expressa como:

$$
f(x) = q \sum_{i,j} \sigma_{ij} x_i x_j - \sum_{i=1}^{N} \mu_i x_i + \gamma \left( \sum_{i=1}^{N} x_i + \sum_{s=0}^{M-1} 2^s y_s - B \right)^2\,,
\tag{8}
$$

onde:
- $x_i \in {0,1}$ indicam a seleção dos ativos (0 - não selecionado; 1 - selecionado);
- $\sigma_{ij}$ representa a matriz de covariância;
- $\mu_i$ os retornos esperados;
- $y_s$ são variáveis auxiliares ("variáveis de folga") para codificação binária da restrição de orçamento ( $B$ );
- $\gamma$ é o coeficiente de penalização.

Essa etapa transforma o problema financeiro em um problema equivalente de minimização de energia, compatível com algoritmos quânticos.

### 6.2 *Ansatz* - Circuito Variacional

Utiliza-se o:
> - `TwoLocal`

Um circuito parametrizado composto por:

- Rotações locais ($R_y$);
- Portas de emaranhamento ($CZ$);
- Profundidade controlada por `reps`.

O parâmetro:

> - `reps = 3`

define a profundidade do circuito, permitindo maior capacidade de representação da solução, ao custo de maior complexidade computacional.

### 6.3 Otimização Híbrida (VQE)

O algoritmo *Variational Quantum Eigensolver* (VQE) combina:

- Circuito quântico parametrizado (*ansatz*);
- Otimizador clássico (COBYLA);
- Simulador quântico:
> - `StatevectorSampler`

O objetivo é minimizar a energia do Hamiltoniano associado ao problema de portfólio, onde a solução final é codificada como um vetor binário indicando os ativos selecionados.

#### 6.3.1 Parametrização do circuito

- $q = 0.5$ <br>
    Controla o trade-off entre risco e retorno. Valores maiores penalizam mais fortemente o risco, enquanto valores menores priorizam retorno.
- `reps = 3`

    Define a profundidade do circuito variacional, impactando diretamente a capacidade de exploração do espaço de soluções.

## 7. Resultados

A saída do modelo é uma solução binária que representa a seleção ótima de ativos segundo o critério definido pelo Hamiltoniano do problema. Inclui-se na análise o gráfico da minimização da energia durante a simulação.

```python
# ==========================================
# Hiperparâmetros
# ==========================================
q = 0.5
budget = 4
maxiter = 400
reps = 4

optimizer = SPSA(maxiter=maxiter)
```

```python
# ==========================================
# Callback para armazenar as energias
# ==========================================
energy_history = []

def store_intermediate_result(eval_count, parameters, value, metadata):
    energy_history.append(value)

# ==========================================
# Modelagem QUBO
# ==========================================
portfolio = PortfolioOptimization(
    expected_returns=mu,
    covariances=sigma,
    risk_factor=q,
    budget=budget
)

qp = portfolio.to_quadratic_program()

# Remove a restrição original "exatamente budget"
qp.remove_linear_constraint(0)

# Adiciona "no máximo budget"
qp.linear_constraint(
    linear=[1] * len(mu),
    sense="<=",
    rhs=budget,
    name="budget_max"
)

print("\n--- Problema QP Gerado ---")
print(qp.prettyprint())

# ==========================================
# Configuração do Algoritmo Quântico (VQE)
# ==========================================
sampler = StatevectorSampler()

ansatz = TwoLocal(
    num_qubits=len(tickers),
    rotation_blocks='ry',
    entanglement_blocks='cz',
    reps=reps,
    entanglement='full'
)

print("\n--- Circuito Variacional (Ansatz) ---\n")
print(ansatz.decompose().draw())

# ==========================================
# Definição do VQE
# ==========================================
if risk_measure == "cvar":
    vqe = SamplingVQE(
        sampler=sampler,
        ansatz=ansatz,
        optimizer=optimizer,
        aggregation=alpha,
        callback=store_intermediate_result
    )
else:
    vqe = SamplingVQE(
        sampler=sampler,
        ansatz=ansatz,
        optimizer=optimizer,
        callback=store_intermediate_result
    )

vqe_optimizer = MinimumEigenOptimizer(vqe)

# ==========================================
# Execução
# ==========================================
print("\nRodando o VQE...")
result = vqe_optimizer.solve(qp)

# ==========================================
# Resultados
# ==========================================
print("\n" + "="*50)
print("Resultado da Otimização Quântica (VQE)")
print("="*50)

print(f"\nSolução binária (x): {result.x}")

selected_assets = [tickers[i] for i, x in enumerate(result.x) if x == 1.0]

x = np.array(result.x)

if x.sum() == 0:
    raise ValueError("Nenhum ativo foi selecionado. Não é possível normalizar os pesos.")

weights = x / x.sum()

portfolio_return = np.dot(weights, mu)

portfolio_risk_value = portfolio_risk(
    weights,
    returns,
    sigma,
    risk_measure=risk_measure,
    alpha=alpha
)

score = portfolio_return / portfolio_risk_value

print(f"\nAtivos selecionados: {selected_assets}")
print(f"\nFunção objetivo: {result.fval:.4f}")
print(f"Retorno esperado: {portfolio_return:.4f}")
print(f"Risco do portfólio: {portfolio_risk_value:.4f}")
print(f"Escore do portfólio (Retorno/Risco): {score:.4f}")

if len(energy_history) > 0:
    print(f"\nÚltima energia registrada: {energy_history[-1]:.6f}")
    print(f"Melhor energia registrada: {min(energy_history):.6f}")
else:
    print("\nNenhuma energia foi registrada no callback.")
```

### 7.1 Comparativo com Método *Brute Force*

Para validar o resultado obtido pelo algoritmo híbrido (VQE), analisamos a carteira selecionada sob a ótica clássica de Finanças Quantitativas, considerando o *trade-off* entre **Risco (Volatilidade)** e **Retorno Esperado**. A análise consiste em:

1. Converter a solução binária em pesos financeiros (cada ativo selecionado recebe 50% do capital);
2. Calcular o retorno esperado do portfólio;
3. Calcular o risco (volatilidade);
4. Plotar todas as combinações no espaço risco-retorno;
5. Destacar a solução encontrada pelo VQE.

Essa abordagem permite verificar se o algoritmo quântico encontrou uma solução próxima da **fronteira eficiente**, ou seja, carteiras que oferecem o melhor retorno possível para um determinado nível de risco. Ao observar o espaço de **Risco vs. Retorno**, podemos tirar conclusões valiosas sobre a performance da solução quântica.

```python
# @title
# 1. Preparar todas as combinações clássicas possíveis (Força Bruta)
num_assets = len(tickers)

# Gerar todas as combinações binárias válidas (onde a soma é igual ao budget)
all_combinations = []
for k in range(1, budget + 1):
    for combo in itertools.combinations(range(num_assets), k):
        x = np.zeros(num_assets)
        x[list(combo)] = 1
        all_combinations.append(x)

returns_list = []
risks_list = []
labels_list = []

# 2. Calcular Risco e Retorno para cada carteira possível
for x in all_combinations:

    # Número de ativos selecionados
    n_selected = np.sum(x)

    # Transformar decisão binária em pesos iguais
    w = x / n_selected

    # Retorno esperado
    port_return = np.dot(w, mu)

    # Risco
    port_risk = portfolio_risk(w, returns, sigma, risk_measure=risk_measure, alpha=alpha)

    returns_list.append(port_return)
    risks_list.append(port_risk)

    # Labels
    assets = [tickers[i] for i in range(num_assets) if x[i] == 1]
    labels_list.append("\n+\n".join(assets))


# 3. Identificar os dados exatos do resultado quântico (VQE)
n_selected_vqe = np.sum(result.x)

w_vqe = result.x / n_selected_vqe

vqe_return = np.dot(w_vqe, mu)
vqe_risk = portfolio_risk(w_vqe, returns, sigma, risk_measure=risk_measure, alpha=alpha)

vqe_label = "+".join(
    [tickers[i] for i in range(num_assets) if result.x[i] == 1]
)

# 4. Plotar o Gráfico de Dispersão (Espaço Risco-Retorno)
plt.figure(figsize=(12, 7))

# Plotar todas as combinações clássicas
plt.scatter(risks_list, returns_list, color='skyblue', s=150, alpha=0.8, edgecolor='black', label='Portfólios Possíveis')

# Anotar os nomes dos ativos perto de cada ponto
for i, label in enumerate(labels_list):
    plt.annotate(label, (risks_list[i], returns_list[i]),
                 xytext=(10, -10), textcoords='offset points',
                 fontsize=9, bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="gray", alpha=0.5))

# Destacar com uma estrela gigante o resultado do VQE
plt.scatter(vqe_risk, vqe_return, color='crimson', marker='*', s=500, edgecolor='black',
            label=f'Seleção Quântica (VQE):\n{vqe_label}', zorder=5)

# Formatação do Gráfico
plt.title('Análise de Risco vs. Retorno: Seleção Quântica vs Todas as Possibilidades', fontsize=16, pad=15)
plt.xlabel('Risco', fontsize=12)
plt.ylabel('Retorno Esperado Anualizado', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(loc= 'best', fontsize=11, framealpha=1)

# Ajuste visual
plt.tight_layout()
plt.show()
```

### 8.1 Refinamento de Hiperparâmetros do modelo quântico binário

```python
# ==========================================
# Função Encapsulada para o Grid Search
# ==========================================

def run_vqe_portfolio(mu, sigma, budget, tickers, risk_factor, penalty, reps, maxiter, risk_measure="variance"):
    # 1. Modelagem (QUBO)
    portfolio = PortfolioOptimization(
        expected_returns=mu,
        covariances=sigma,
        risk_factor=risk_factor,
        budget=budget
    )

    qp = portfolio.to_quadratic_program()
    qp.remove_linear_constraint(0) # Remove "exatamente budget"
    qp.linear_constraint(          # Adiciona "no máximo budget"
        linear=[1] * len(mu),
        sense="<=",
        rhs=budget,
        name="budget_max"
    )

    # 2. Configuração Quântica (Ansatz, Otimizador, VQE)
    sampler = StatevectorSampler()

    ansatz = TwoLocal(
        num_qubits=len(tickers),
        rotation_blocks='ry',
        entanglement_blocks='cz',
        reps=reps,
        entanglement='full'
    )

    optimizer = SPSA(maxiter=maxiter)

    if risk_measure == "cvar":
        vqe = SamplingVQE(sampler=sampler, ansatz=ansatz, optimizer=optimizer, aggregation=alpha)
    else:
        vqe = SamplingVQE(sampler=sampler, ansatz=ansatz, optimizer=optimizer)

    # Nota: passamos o 'penalty' diretamente para o EigenOptimizer
    # Se penalty=None, o Qiskit calcula a penalidade das restrições automaticamente.
    vqe_optimizer = MinimumEigenOptimizer(vqe, penalty=penalty)

    # 3. Execução
    result = vqe_optimizer.solve(qp)

    # Extração de ativos selecionados
    selected_assets = [tickers[i] for i, x in enumerate(result.x) if x == 1.0]

    return {
        "x": result.x,
        "selected_assets": selected_assets,
        "fval": result.fval,
        "status": result.status.name
    }

# ==========================================
# Configuração e Execução do Grid Search
# ==========================================

# Hiperparâmetros a serem testados no refinamento
q_values = [0.5, 1.0, 2.0]       # Fator de aversão ao risco (substitui o seu fator q original)
penalty_values = [None, 5, 10]   # Penalidade para violação de restrições (None = auto)
reps_list = [2, 3, 4]            # Profundidade do ansatz variacional
maxiters = [400]       # Número máximo de iterações do otimizador

results = []

# Iteramos sobre todas as combinações possíveis
for q, penalty, reps, maxiter in itertools.product(q_values, penalty_values, reps_list, maxiters):

    try:
        # Chama a função encapsulada
        res = run_vqe_portfolio(
            mu=mu,
            sigma=sigma,
            budget=budget,
            tickers=tickers,
            risk_factor=q,
            penalty=penalty,
            reps=reps,
            maxiter=maxiter,
            risk_measure=risk_measure # Assume-se que 'risk_measure' foi definida globalmente
        )

        results.append({
            "q (risk)": q,
            "penalty": penalty,
            "reps": reps,
            "maxiter": maxiter,
            "fval": res["fval"],
            "ativos": ", ".join(res["selected_assets"]),
            "status": res["status"]
        })

    except Exception as e:
        print(f"Erro na combinação: {e}")
        # Para o loop não quebrar em caso de falha pontual
        results.append({
            "q (risk)": q, "penalty": penalty, "reps": reps, "maxiter": maxiter,
            "fval": None, "ativos": "ERRO", "status": "FAIL"
        })

# ==========================================
# Consolidação dos Resultados
# ==========================================

df = pd.DataFrame(results)

# O PortfolioOptimization do Qiskit formula um problema de MINIMIZAÇÃO (Risco - Retorno).
# Portanto, usamos ascending = True.
df_sorted = df.sort_values("fval", ascending=True)

print("\n" + "="*80)
print("Resultados do Grid Search VQE (Ordenados do Melhor para o Pior FVAL)")
print("="*80)
print(df_sorted.to_string(index=False))
```

### 8.2 Otimização de porcentagens (não-binária) usando Pauli Correlator Encoding (PCE)

Nesta seção, substituímos a formulação binária (comprar ou não comprar uma ação) por uma **alocação percentual contínua** utilizando **Pauli Correlator Encoding (PCE)** dentro de um **Variational Quantum Eigensolver (VQE)**.

Essa abordagem permite construir um portfólio mais realista, onde cada ativo recebe uma **porcentagem da carteira**, em vez de uma decisão binária.

---

### Ideia Geral

Anteriormente, o modelo utilizava variáveis binárias:

$$x_i \in \{0,1\}$$

Onde:
* $x_i = 1$ → comprar ativo
* $x_i = 0$ → não comprar

Agora, utilizamos variáveis contínuas:

$$w_i \in [0,1]$$

com a restrição:

$$\sum_i w_i = 1$$

onde:
* $w_i$ representa a **porcentagem do capital alocada no ativo $i$**

---

### Pauli Correlator Encoding (PCE)

No PCE, cada ativo é associado a um **correlator de operadores de Pauli**:

Exemplos:
* $X_0 X_1$
* $Y_1 Y_2$
* $Z_0 Z_3$

Cada correlator é um **observável quântico**:

$$P_i$$

Após a execução do circuito variacional, calculamos:

$$\langle P_i \rangle = \langle \psi(\theta) | P_i | \psi(\theta) \rangle$$

Esses valores pertencem ao intervalo:

$$[-1, 1]$$

Esses números são interpretados como **scores quânticos dos ativos**.

---

### Conversão para Pesos do Portfólio

Os valores esperados são convertidos em pesos utilizando **Softmax**:

$$w_i = \frac{e^{x_i/T}}{\sum_j e^{x_j/T}}$$

Onde:
* $T$ é a temperatura
* Controla a diversificação:
    * **Temperatura alta** → mais diversificação
    * **Temperatura baixa** → maior concentração

Isso garante:
* Pesos positivos
* Soma igual a 1
* Interpretação como porcentagem do portfólio

---

### Circuito Variacional

O circuito utilizado é um **TwoLocal ansatz**:
* Rotações: $R_y$
* Emaranhamento: $CZ$
* Múltiplas camadas

O estado quântico produzido é:

$$|\psi(\theta)\rangle$$

Os parâmetros variacionais $\theta$ são otimizados para encontrar o melhor portfólio.

---

### Função Objetivo

A função objetivo otimiza simultaneamente:

#### Retorno esperado

$$R = \sum_i w_i \mu_i$$

onde:
* $\mu_i$ é o retorno esperado do ativo

---

#### Risco do portfólio

Dependendo da configuração, o risco pode ser Variância, CVaR ou outra métrica implementada:

$$\sigma = w^T \Sigma w$$

---

#### Penalidade de concentração

Para evitar concentração excessiva:

$$Penalty = \sum_i w_i^2$$

Isso incentiva a diversificação.

---

### Função de Loss

A função final otimizada é:

$$L = q \cdot \text{Risco} - \text{Retorno} + \lambda \cdot \text{Penalidade}$$

Onde:
* $q$ controla aversão ao risco
* $\lambda$ controla diversificação

---

### Otimização

Utilizamos um otimizador clássico para encontrar:

$$\theta^*$$

que minimiza a função objetivo.

---

### Resultado Final

Após a otimização:
1.  Calculamos os *expectation values*
2.  Aplicamos softmax
3.  Obtemos pesos percentuais

**Exemplo de Portfólio:**

| Ativo | Peso |
| :--- | :--- |
| PETR4 | 22% |
| VALE3 | 18% |
| ITUB4 | 15% |
| ABEV3 | 12% |

Isso gera um portfólio contínuo e realista.

---

### Vantagens do Pauli Correlator Encoding

* Melhor uso de poucos qubits
* Alocação contínua
* Portfólio mais realista
* Melhor diversificação
* Totalmente compatível com VQE

```python
# ==========================================
# Hiperparâmetros
# ==========================================

reps = 3
maxiter = 400
temperature = 1.15 # maior -> alocação mais uniforme
alpha = 0.05
risk_measure = "cvar"
q = 0.5
diversification_penalty = 0.02
```

```python
# =========================================================
# Pauli Correlator Encoding (PCE) -> pesos percentuais
# =========================================================

def softmax(x, temperature=1.0):
    x = np.asarray(x, dtype=float)
    temperature = max(float(temperature), 1e-12)
    z = x / temperature
    z = z - np.max(z)
    e = np.exp(z)
    s = e.sum()
    return e / s if s > 0 else np.ones_like(e) / len(e)

def min_qubits_for_assets(n_assets, k=2, families=3):
    q = max(k, 2)
    while families * math.comb(q, k) < n_assets:
        q += 1
    return q

def build_pauli_correlation_encoding(asset_names, n_qubits, k=2):
    """
    Mapeia cada ativo para uma string de Pauli de peso k.
    Usamos famílias X/Y/Z em combinações distintas de qubits.
    """
    pairs = list(itertools.combinations(range(n_qubits), k))
    families = ["X", "Y", "Z"]

    asset_to_op = {}
    asset_to_tag = {}

    idx = 0
    for family in families:
        for pair in pairs:
            if idx >= len(asset_names):
                break

            paulis = ["I"] * n_qubits
            for q in pair:
                paulis[q] = family

            # Qiskit usa a convenção invertida na string
            pauli_string = "".join(paulis)[::-1]

            asset_to_op[asset_names[idx]] = SparsePauliOp.from_list([(pauli_string, 1.0)])
            asset_to_tag[asset_names[idx]] = f"{family}{pair[0]}{pair[1]}"
            idx += 1

        if idx >= len(asset_names):
            break

    return asset_to_op, asset_to_tag

# =========================================================
# 1) Preparação do encoding
# =========================================================

asset_names = list(tickers)
n_assets = len(asset_names)

# Escolha compacta de qubits: suficientes para 3 famílias de correlatores 2-body
num_qubits_pce = min_qubits_for_assets(n_assets, k=2, families=3)

asset_ops, asset_tags = build_pauli_correlation_encoding(
    asset_names=asset_names,
    n_qubits=num_qubits_pce,
    k=2
)

print(f"Ativos: {n_assets} | Qubits PCE: {num_qubits_pce}")
print("Mapeamento ativo -> correlator:")
for t in asset_names:
    print(f"  {t}: {asset_tags[t]}")

# =========================================================
# 2) Ansatz variacional
# =========================================================

ansatz = TwoLocal(
    num_qubits=num_qubits_pce,
    rotation_blocks="ry",
    entanglement_blocks="cz",
    reps=reps,
    entanglement="full"
)

print("\nCircuito variacional (ansatz) criado com sucesso.")

# =========================================================
# 3) Decodificação: expectativas -> porcentagens
# =========================================================

def get_expectations(theta):
    bound_circuit = ansatz.assign_parameters(theta)
    state = Statevector.from_instruction(bound_circuit)
    exp_vals = np.array([
        float(np.real(state.expectation_value(asset_ops[t])))
        for t in asset_names
    ], dtype=float)
    return exp_vals

def decode_weights(exp_vals, temperature=1.35):
    # Softmax garante pesos positivos somando 1
    return softmax(exp_vals, temperature=temperature)

def portfolio_objective(theta):
    exp_vals = get_expectations(theta)
    weights = decode_weights(exp_vals, temperature=temperature)

    port_return = float(np.dot(weights, mu))
    port_risk = float(portfolio_risk(weights, returns, sigma, risk_measure=risk_measure, alpha=alpha))
    concentration = float(np.sum(weights ** 2))  # penaliza excesso de concentração

    # Minimiza risco e concentração, maximiza retorno
    loss = q * port_risk - port_return + diversification_penalty * concentration
    return loss

# =========================================================
# 4) Otimização clássica sobre o circuito quântico
# =========================================================

rng = np.random.default_rng(42)
x0 = rng.uniform(0, 2*np.pi, ansatz.num_parameters)

print("\nOtimizando parâmetros variacionais...")
opt_result = minimize(
    portfolio_objective,
    x0=x0,
    method="COBYLA",
    options={"maxiter": maxiter, "disp": True, "rhobeg": 0.5}
)

best_theta = opt_result.x
best_exp_vals = get_expectations(best_theta)
best_weights = decode_weights(best_exp_vals, temperature=temperature)

# =========================================================
# 5) Resultados finais
# =========================================================

allocation_df = pd.DataFrame({
    "Ticker": asset_names,
    "Correlator": [asset_tags[t] for t in asset_names],
    "Pauli Exp.": best_exp_vals,
    "Peso": best_weights,
    "Percentual": 100 * best_weights,
    "Retorno esp.": mu
}).sort_values("Percentual", ascending=False).reset_index(drop=True)

portfolio_expected_return = float(np.dot(best_weights, mu))
portfolio_expected_risk = float(portfolio_risk(best_weights, returns, sigma, risk_measure=risk_measure, alpha=alpha))

selected_assets = allocation_df.loc[allocation_df["Percentual"] >= 5.0, "Ticker"].tolist()

print("\n" + "="*70)
print("ALOCAÇÃO FINAL (PCE)")
print("="*70)
print(allocation_df.to_string(
    index=False,
    formatters={
        "Pauli Exp.": "{:.6f}".format,
        "Peso": "{:.6f}".format,
        "Percentual": "{:.2f}%".format,
        "Retorno esp.": "{:.4f}".format,
    }
))

print("\n" + "="*70)
print("Resumo da carteira")
print("="*70)
print(f"Soma dos pesos: {best_weights.sum():.6f}")
print(f"Retorno esperado do portfólio: {portfolio_expected_return:.4f}")
print(f"Risco do portfólio: {portfolio_expected_risk:.4f}")
print(f"Escore do portfólio: {(portfolio_expected_return/portfolio_expected_risk):.4f}")
print(f"Status do otimizador: {opt_result.message}")

# =========================================================
# 6) Gráfico das porcentagens
# =========================================================

plt.figure(figsize=(12, 6))
plt.bar(allocation_df["Ticker"], allocation_df["Percentual"])
plt.title("Alocação percentual via Pauli Correlator Encoding")
plt.xlabel("Ativo")
plt.ylabel("Percentual da carteira (%)")
plt.grid(axis="y", linestyle="--", alpha=0.4)
plt.tight_layout()
plt.show()

# Compatibilidade com trechos posteriores do notebook
result = opt_result
w_pce = best_weights
```

```python
# =========================================================
# 1) Preparação do encoding
# =========================================================

asset_names = list(tickers)
n_assets = len(asset_names)

# Mantém fixo o número de qubits para o experimento
num_qubits_pce = min_qubits_for_assets(n_assets, k=2, families=3)

asset_ops, asset_tags = build_pauli_correlation_encoding(
    asset_names=asset_names,
    n_qubits=num_qubits_pce,
    k=2
)

print(f"Ativos: {n_assets} | Qubits PCE: {num_qubits_pce}")
print("Mapeamento ativo -> correlator:")
for t in asset_names:
    print(f"  {t}: {asset_tags[t]}")

# =========================================================
# 2) Função encapsulada para rodar o PCE
# =========================================================

def run_pce_portfolio(
    mu,
    sigma,
    returns_df,
    tickers,
    q,
    reps,
    maxiter,
    temperature,
    diversification_penalty,
    alpha=0.05,
    risk_measure="variance",
    seed=42,
):
    """
    Executa a otimização variacional do portfólio via PCE
    e retorna os principais resultados.
    """

    # Ansatz variacional
    ansatz = TwoLocal(
        num_qubits=num_qubits_pce,
        rotation_blocks="ry",
        entanglement_blocks="cz",
        reps=reps,
        entanglement="full"
    )

    def get_expectations(theta):
        bound_circuit = ansatz.assign_parameters(theta)
        state = Statevector.from_instruction(bound_circuit)
        exp_vals = np.array([
            float(np.real(state.expectation_value(asset_ops[t])))
            for t in asset_names
        ], dtype=float)
        return exp_vals

    def decode_weights(exp_vals):
        return softmax(exp_vals, temperature=temperature)

    def portfolio_objective(theta):
        exp_vals = get_expectations(theta)
        weights = decode_weights(exp_vals)

        port_return = float(np.dot(weights, mu))
        port_risk = float(portfolio_risk(weights, returns_df, sigma, risk_measure=risk_measure, alpha=alpha))
        concentration = float(np.sum(weights ** 2))

        # Minimiza risco e concentração, maximiza retorno
        loss = q * port_risk - port_return + diversification_penalty * concentration
        return loss

    rng = np.random.default_rng(seed)
    x0 = rng.uniform(0, 2 * np.pi, ansatz.num_parameters)

    opt_result = minimize(
        portfolio_objective,
        x0=x0,
        method="COBYLA",
        options={"maxiter": maxiter, "disp": False, "rhobeg": 0.5}
    )

    best_theta = opt_result.x
    best_exp_vals = get_expectations(best_theta)
    best_weights = decode_weights(best_exp_vals)

    allocation_df = pd.DataFrame({
        "Ticker": asset_names,
        "Correlator": [asset_tags[t] for t in asset_names],
        "Pauli Exp.": best_exp_vals,
        "Peso": best_weights,
        "Percentual": 100 * best_weights,
        "Retorno esp.": mu
    }).sort_values("Percentual", ascending=False).reset_index(drop=True)

    portfolio_expected_return = float(np.dot(best_weights, mu))
    portfolio_expected_risk = float(portfolio_risk(best_weights, returns_df, sigma, risk_measure=risk_measure, alpha=alpha))
    score = portfolio_expected_return / portfolio_expected_risk if portfolio_expected_risk != 0 else np.nan

    return {
        "x": best_theta,
        "allocation_df": allocation_df,
        "weights": best_weights,
        "exp_vals": best_exp_vals,
        "portfolio_expected_return": portfolio_expected_return,
        "portfolio_expected_risk": portfolio_expected_risk,
        "score": score,
        "status": opt_result.message,
        "success": opt_result.success,
        "fun": opt_result.fun,
        "nfev": opt_result.nfev,
        "reps": reps,
        "maxiter": maxiter,
        "q": q,
        "temperature": temperature,
        "diversification_penalty": diversification_penalty,
    }

# =========================================================
# 3) Grid search dos hiperparâmetros
# =========================================================

q_values = [0.5, 1.0]
temperature_values = [1.0, 1,15, 1.35]
diversification_penalties = [0.0, 0.01, 0.02, 0.05]
reps_list = [2, 3, 4]
maxiters = [100, 200, 300]

results = []

for q, temperature, diversification_penalty, reps, maxiter in itertools.product(
    q_values,
    temperature_values,
    diversification_penalties,
    reps_list,
    maxiters
):
    try:
        res = run_pce_portfolio(
            mu=mu,
            sigma=sigma,
            returns_df=returns,
            tickers=tickers,
            q=q,
            reps=reps,
            maxiter=maxiter,
            temperature=temperature,
            diversification_penalty=diversification_penalty,
            alpha=alpha,
            risk_measure=risk_measure,
            seed=42
        )

        results.append({
            "q": q,
            "temperature": temperature,
            "diversification_penalty": diversification_penalty,
            "reps": reps,
            "maxiter": maxiter,
            "fun": res["fun"],
            "return": res["portfolio_expected_return"],
            "risk": res["portfolio_expected_risk"],
            "score": res["score"],
            "status": res["status"],
            "success": res["success"],
            "nfev": res["nfev"],
        })

    except Exception as e:
        print(f"Erro na combinação (q={q}, temp={temperature}, pen={diversification_penalty}, reps={reps}, maxiter={maxiter}): {e}")
        results.append({
            "q": q,
            "temperature": temperature,
            "diversification_penalty": diversification_penalty,
            "reps": reps,
            "maxiter": maxiter,
            "fun": None,
            "return": None,
            "risk": None,
            "score": None,
            "ativos": "ERRO",
            "status": "FAIL",
            "success": False,
            "nfev": None,
        })

# =========================================================
# 4) Consolidação dos resultados
# =========================================================

df = pd.DataFrame(results)

# Como o objetivo é minimização, o melhor resultado tem menor valor de fun
df_sorted = df.sort_values("fun", ascending=True)

print("\n" + "=" * 90)
print("Resultados do Grid Search PCE (ordenados do melhor para o pior FUN)")
print("=" * 90)
print(df_sorted.to_string(index=False))

# Melhor combinação
best_row = df_sorted.iloc[0]
print("\n" + "=" * 90)
print("Melhor combinação encontrada")
print("=" * 90)
print(best_row.to_string())
```
